```python
import yaml

with open('Custom_data.yaml', 'r') as f:
    data_yaml = yaml.safe_load(f)

# Correct the paths
data_yaml['train'] = data_yaml['train'].replace('TrainingData/', '')
data_yaml['val'] = data_yaml['val'].replace('TrainingData/', '')

with open('Custom_data.yaml', 'w') as f:
    yaml.dump(data_yaml, f)

print("Updated Custom_data.yaml:")
print(data_yaml)
```

```python
model = YOLO("yolov8n.pt")

model.train(
    data="Custom_data.yaml",
    epochs=50,
    imgsz=320,
    batch=16
)
```
